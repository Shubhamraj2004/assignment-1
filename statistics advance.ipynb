Question 2. what are the types of random variable.
Answer. There are two main types of random variables:
Discrete Random Variable: A discrete random variable can take on a finite or countably infinite number of values. These values are typically integers and represent counts or categories.
Examples: The number of heads in 10 coin flips, the number of defective items in a sample, the number of cars passing a certain point in an hour.
Continuous Random Variable: A continuous random variable can take on any value within a given range or interval. These values are typically measurements.
Examples: The height of a person, the temperature of a room, the time it takes to complete a task.
Question 3. what is the difference between dicrete and continuous distribution.
Answer. Discrete Distributions: Describe the probabilities of discrete random variables. The probability is assigned to specific, distinct values. The sum of all probabilities for all possible values must equal 1.
Example: The probability of getting exactly 3 heads in 5 coin flips.
Continuous Distributions: Describe the probabilities of continuous random variables. Since there are infinitely many possible values, the probability of a continuous random variable taking on any single specific value is essentially zero. Instead, probabilities are calculated over intervals of values. The total area under the probability density curve must equal 1.
Example: The probability that a person's height is between 170 cm and 175 cm.
Question 4.what are probability distribution function.
Answer. The term "probability distribution function" can be a bit ambiguous as it's sometimes used generally. However, more precisely:
For discrete random variables, it's called a Probability Mass Function (PMF). A PMF gives the probability that a discrete random variable is exactly equal to some value.
Notation: P(X=x) or f(x)
For continuous random variables, it's called a Probability Density Function (PDF). A PDF does not give the probability of a specific value. Instead, the area under the curve of the PDF between two points gives the probability that the random variable falls within that interval.
Notation: f(x)

CDF: Gives the cumulative probability up to a specific point. It always starts at 0 and ends at 1.

Question 13.what is the central limit theorem and why is it critical in statistics.
Answer. 
The Central Limit Theorem (CLT) is a fundamental theorem in probability theory. It states that, given a sufficiently large sample size from any population distribution with a finite mean and variance, the distribution of the sample means will approximate a normal distribution, regardless of the shape of the original population distribution.
Criticality in Statistics:
Foundation for Inferential Statistics: The CLT is the cornerstone of many statistical inferene techniques, particularly hypothesis testing and confidence intervals.
Normality Assumption: Even if the underlying population data is not normally distributed, the CLT allows us to use normal distribution theory to make inferences about population parameters (like the mean) as long as our sample size is large enough.
Approximation: It explains why the normal distribution is so prevalent in real-world data analysis, even when the original data doesn't appear normal. Many real-world phenomena are the result of summing or averaging many small, independent random effects.
Sample Size: The "sufficiently large" sample size typically refers to n≥30, though this can vary depending on the skewness of the original distribution.

Question 15.what is application of z statistics in hypothesis testing.
Answer. 
Standardization: Z-statistics standardize the sample statistic (e.g., sample mean, sample proportion) by converting it into a number of standard deviations it is away from the hypothesized population parameter under the null hypothesis. The formula for a Z-statistic 
Determining p-value: Once the Z-statistic is calculated, it is compared to a standard normal distribution (Z-table) to find the p-value. The p-value is the probability of observing a sample statistic as extreme as, or more extreme than, the one observed, assuming the null hypothesis is true.
Decision Making:
If the p-value is less than the chosen significance level (α), we reject the null hypothesis. This means the observed difference is statistically significant and unlikely to have occurred by chance.
If the p-value is greater than α, we fail to reject the null hypothesis. This means there isn't enough evidence to conclude a significant difference.
Critical Value Approach: Alternatively, the calculated Z-statistic can be compared to critical Z-values (obtained from the Z-table for a given α). If the calculated Z-statistic falls into the rejection region (beyond the critical value), the null hypothesis is rejected.

Question 18. what is the significance of confindence intervals in statistical analysis.
Answer. Quantifying Uncertainty: They provide a measure of the uncertainty or precision associated with a sample-based estimate of a population parameter. A narrower interval indicates a more precise estimate.

Inference about Population Parameters: They allow us to make inferences about the true value of a population parameter based on sample data. If a specific hypothesized value for a parameter falls outside the confidence interval, it provides strong evidence against that hypothesis.

Decision Making: In many practical applications (e.g., medical research, quality control, marketing), confidence intervals help in making informed decisions. For example, if a new drug's effect has a confidence interval that includes zero, it might suggest no significant effect.

Question 21.what are the assumption for applying the central limit theorem .
Answer.Random Sampling: The samples must be drawn randomly from the population. This ensures that the samples are representative and reduces bias.
Independence: The individual observations within each sample must be independent of each other. Also, if sampling without replacement, the population must be much larger than the sample size (typically, population size ≥20× sample size) to ensure approximate independence between samples.
Sample Size is Sufficiently Large: This is the most crucial assumption. There's no strict rule for "sufficiently large," but a general guideline is n≥30. For populations that are highly skewed or have unusual shapes, a larger sample size might be needed for the sampling distribution to be approximately normal. If the population itself is normally distributed, then the sampling distribution of the mean will be normal for any sample size.
Finite Mean and Variance: The population from which the samples are drawn must have a finite mean (μ) and a finite variance (σ 2). This is generally true for most real-world populations.
Question 22. what is the concept of expected value in probability distribution.
Answer. The expected value (or expectation, denoted as E[X]) of a random variable X in a probability distribution is the weighted average of all possible values that the random variable can take, where the weights are the probabilities of those values occurring. It represents the long-run average or the theoretical mean of the distribution.
Question 23.how does a probability distribution relate to the expected outcome of a random variable.
Answer. The Probability Distribution Defines the Possibilities: The probability distribution lays out the complete picture of uncertainty for a random variable. It tells you what values are possible and how likely each value is.
The Expected Outcome is a Summary Statistic: The expected value is derived directly from the probability distribution. It's calculated by taking all the possible outcomes defined by the distribution and weighting them by their respective probabilities.

Theoretical Long-Run Average: The expected value is the theoretical long-run average of the random variable if the experiment were to be repeated an infinite number of times. It's the outcome you would "expect" to see, on average, over many trials.

Measure of Central Tendency: For a probability distribution, the expected value serves as its mean. It indicates the central location around which the values of the random variable tend to cluster.
